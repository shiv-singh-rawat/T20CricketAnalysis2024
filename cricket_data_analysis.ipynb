```python
import pandas as pd
```

```python
batting_df = pd.read_csv('/content/Batting summaries for every match.csv')
```

```python
batting_df.head(10)
```

```python
highest_score = batting_df.groupby(['batsmanName', 'Match_Id'])['runs'].sum().reset_index()
```

```python
highest_score.sort_values(by='runs', ascending=False)
```

```python
highest_score = batting_df.groupby(['batsmanName', 'Match_Id']).sum()['runs'].reset_index()
```

```python
highest_score.sort_values(by='runs', ascending=False).head(10)
```

```python
# Group by 'batsmanName' to calculate total runs scored by each player across all matches
total_runs_per_player = batting_df.groupby('batsmanName')['runs'].sum().reset_index()
```

```python
# Sort the players by total runs in descending order to find the player with the most runs
total_runs_per_player_sorted = total_runs_per_player.sort_values(by='runs', ascending=False)
```

```python
# To see the result (most runs)
total_runs_per_player_sorted.head(10)
```

```python
best_batting_avg = batting_df.groupby('batsmanName')['runs'].mean().reset_index()
best_batting_avg.sort_values(by='runs', ascending=False).head(10)
```

```python
# Group by 'batsmanName' and calculate the average runs for each player
average_runs_per_player = batting_df.groupby('batsmanName')['runs'].mean().reset_index()

# To see the result
average_runs_per_player.sort_values(by='runs', ascending=False)
```

```python
total_fours_by_player = batting_df.groupby('batsmanName')['4s'].sum().reset_index(name='total_fours')

# Sort the players by the total number of fours in descending order
most_fours_player = total_fours_by_player.sort_values(by='total_fours', ascending=False)
```

```python
# Display the player with the most fours
most_fours_player.head(10)
```

```python
total_sixes_by_player = batting_df.groupby('batsmanName')['6s'].sum().reset_index(name='total_fours')

# Sort the players by the total number of fours in descending order
most_fours_player = total_sixes_by_player.sort_values(by='total_fours', ascending=False)
```

```python
# Display the player with the most fours
most_fours_player.head(10)
```

```python
performance_by_position = batting_df.groupby('battingPos').agg(
    total_runs=('runs', 'sum'),
    total_boundaries=('4s', 'sum'),  # You can add '6s' if it's a separate column
    total_sixes=('6s', 'sum'),       # If you have the '6s' column
    average_sr=('SR', 'mean')
).reset_index()
```

```python
print(performance_by_position)
```

```python
performance_by_out_status = batting_df.groupby('out/not_out').agg(
    total_runs=('runs', 'sum'),
    total_balls_faced=('balls', 'sum'),
    average_sr=('SR', 'mean'),
    total_boundaries=('4s', 'sum'),  # Adjust for your columns
    total_sixes=('6s', 'sum')
).reset_index()
```
