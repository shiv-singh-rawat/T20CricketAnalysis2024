```python
import pandas as pd
```

```python
bowling_df = pd.read_csv('/content/Bowling summaries for every match.csv')
```

```python
bowling_df.head(10)
```

### ***Most Wickets***

```python
most_wicket = bowling_df.groupby('bowlerName')['wickets'].sum().reset_index()
```

```python
most_wicket = most_wicket.sort_values(by='wickets', ascending=False)
```

```python
most_wicket.head(10)
```

### ***Best Economy Stats***

```python
# Calculate economy rate for each bowler directly using groupby and aggregate
economy_stats = (
    bowling_df.groupby('bowlerName', as_index=False)
    .agg(total_runs=('runsConceded', 'sum'), total_overs=('overs', 'sum'))
)
```

```python
# Add a new column for economy rate
economy_stats['economy_rate'] = economy_stats['total_runs'] / economy_stats['total_overs']
```

```python
# Filter out invalid data (if any bowler has zero overs)
economy_stats = economy_stats[economy_stats['total_overs'] > 0]
```

```python

# Sort by economy rate in ascending order
economy_stats['economy_rate'] = economy_stats['economy_rate'].round(2)
economy_stats = economy_stats.sort_values(by='economy_rate', ascending=True)

economy_stats[['bowlerName', 'economy_rate']].head(10) # Display the top 10 bowlers
```

```python
tim_southee_stats = bowling_df[bowling_df['bowlerName'] == 'Tim Southee']
```

```python
tim_southee_summary = tim_southee_stats.groupby('bowlerName').agg(
    total_overs = ('overs','sum'),
    total_runs_conceded = ('runsConceded','sum')
).reset_index()
```

```python
tim_southee_summary
```

### Best Bowling **Avg** italicized text

```python
bowling_avg = bowling_df.groupby('bowlerName').apply(
    lambda x: x['runsConceded'].sum() / x['wickets'].sum() if x['wickets'].sum() > 0 else None
).reset_index(name='bowling_avg')
```

```python
bowling_avg.sort_values(by='bowling_avg', ascending=True).head(10).round(2)
```
